<h3>Import Libraries</h3>

```python
# Libraries data analysis
import numpy as np
import pandas as pd

print('Libraries imported!')
```

<p>We obtained the following Dataframe after scrapping the data from the <A href='https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=945633050'><u>wikipedia</u></A> page and cleanning it. We stored the cleaned data in 'toronto_cleaned.csv'</p>

```python
# import the dataset and have a look at our data.
toronto_df = pd.read_csv('toronto_cleaned.csv')
toronto_df.head(10)
```

<h3>Import the coordinates from the 'Geospatial_Coordinates.csv' file</h3>

```python
# load the coordinates from the csv file on Coursera
toronto_coordinates = pd.read_csv("Geospatial_Coordinates.csv")
toronto_coordinates.head(10)
```

<h4>Rename columns as required</h4>
We rename 'Postal Code' column to 'PostalCode'

```python
# Let's rename the column "Postal Code" to "PostalCode"
toronto_coordinates.rename(columns= {'Postal Code':'PostalCode'}, inplace= True)
toronto_coordinates.head(3)
```

<h3>Merge both the Dataframes to obtain the required Dataframe</h3>

```python
# let's merge both the dataframe on the column "PostalCode"
toronto_df_updated = toronto_df.merge(toronto_coordinates, on= 'PostalCode', how= 'left')
toronto_df_updated.head(10)
```

<h3>Let's Check to make sure the coordinates are added as required</h3>

```python
column_names = ["PostalCode", "Borough", "Neighborhood", "Latitude", "Longitude"]
req_dataframe = pd.DataFrame(columns= column_names)

req_postalcodes = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for postcode in req_postalcodes:
    req_dataframe = req_dataframe.append(toronto_df_updated[toronto_df_updated["PostalCode"] == postcode], ignore_index=True)
    
req_dataframe
```

### Convert updated dataframe to CSV file

```python
#Converting obtained DataFrame to CSV file
toronto_df_updated.to_csv('toronto_updated.csv', index=False)
```
